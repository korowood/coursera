```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
mu = 2.0
sigma = 0.5
norm_rv = sts.norm(loc=mu, scale=sigma)
vib = norm_rv.rvs(1000)
```

```python
x = np.linspace(0, 4, 100)
# плотность распределения
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label='Теоретическая плотность распределения', color='black')
# гистограмма выборки
plt.hist(vib, bins=40, density = True, label = 'Прогнозируемая плотность распределения', color = 'red')
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.title('Нормальное распределение')
plt.xlim((0, 4))
plt.legend(loc='upper left')
```

```python
steps = [1, 5, 10, 20]

for n in steps:
    mylist = []
    for i in range(1000):
        mylist.append(np.mean(norm_rv.rvs(n)))
    mu = 2.0 # мат ожидание
    disp = sigma ** 2 #дисперсия распределения
    x = np.linspace(0, 4, 100)
    norm_rv2 = sts.norm(mu, sigma**2)
    plt.hist(mylist,bins=20,density=True,label='Sample mean', color='red')
    plt.plot(x,norm_rv2.pdf(x), color='black')
    plt.ylabel('$f(x)$')
    plt.xlabel('$x$')
    plt.title('n = %d'%n)
    plt.legend(loc='upper right')
    plt.show()
```

#### Дубль 2

```python
scale = 2
exp_rv = sts.expon(scale=scale)

vib = exp_rv.rvs(1000)
```

```python
x = np.linspace(0, 10, 1000)
# плотность распределения
pdf = exp_rv.pdf(x)
plt.plot(x, pdf, label='Теоретическая плотность распределения', color='black')
# гистограмма выборки
plt.hist(vib, bins=40, density = True, label = 'Прогнозируемая плотность распределения', color = 'red')
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.title('Экспонециальное распределение')
plt.xlim((0, 4))
plt.legend(loc='upper right')
```

```python
steps = [1, 5, 10, 20, 100]

for n in steps:
    mylist2 = []
    for i in range(1000):
        mylist2.append(np.mean(exp_rv.rvs(n)))
    E = scale # мат ожидание
    disp = (scale ** 2) #дисперсия распределения
    x = np.linspace(-2, 8, 1000)
    exp_rv2 = sts.expon(scale=scale)
    plt.hist(mylist2,bins=20,density=True,label='Sample mean', color='red')
    mu = E 
    sigma = (disp /n) ** 0.5
    norm_rv = sts.norm(loc=mu, scale=sigma)
    plt.plot(x, norm_rv.pdf(x), color='black')
    plt.ylabel('$f(x)$')
    plt.xlabel('$x$')
    plt.title('n = %d'%n)
    plt.legend(loc='upper right')
    plt.show()
```

Уже лучше, но все равно normed = True ругается.
При малых размерах выборки n нормальное распределение из ЦПТ плохо аппроксимировало распределение выборочных средних. При увеличении n распределение всё больше походило на нормальное и его дисперсия уменьшалась. Точность аппроксимации нормальным распределением увеличивается с ростом n
